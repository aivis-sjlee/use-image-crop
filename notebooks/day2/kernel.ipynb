```python
import torch
import torch.nn as nn

# 1. 128x128 크기의 랜덤한 3채널(RGB) 이미지가 들어온다고 칩시다.
# (Batch_size, Channel, Height, Width)
input_image = torch.randn(1, 3, 128, 128)

# 2. 커널(CNN) 레이어를 만듭니다.
conv_layer = nn.Conv2d(
    in_channels=3,  # 입력 채널 (RGB)
    out_channels=16, # 커널의 개수 (16개의 서로 다른 돋보기로 훑겠다!)
    kernel_size=3,   # 커널의 크기 (3x3 돋보기)
    padding=1        # 이미지 가장자리도 잘 훑기 위한 여백
)
model = nn.Sequential(
    # 1층
    nn.Conv2d(3, 16, kernel_size=3, padding=1), # 16개의 기초 특징 산출
    nn.ReLU(),
    nn.MaxPool2d(kernel_size=2, stride=2), # 128 & 128 -> 64 & 64

    # 2층
    nn.Conv2d(16, 32, kernel_size=3, padding=1), #기초 특징 16개로 복잡한 32개의 특징 산출
    nn.ReLU(),
    nn.MaxPool2d(kernel_size=2, stride=2), # 64 & 64 -> 32 & 32
)
output = model(input_image)

print(f"원본 이미지 크기: {input_image.shape}")
print(f"2층 통과 후 크기: {output.shape}")
```
